```python
#IMPORTAR AS BIBLIOTECAS:
import pandas as pd
import itertools
```

```python
#Criar DataFrame Sorteios passados
df= pd.read_excel('/content/Lotofacil (3).xlsx')
df.head()
```

```python
# Gerar todas as combinações possíveis da Lotofácil (de 1 a 25)
numeros_lotofacil = list(range(1, 26))
todas_combinacoes = list(itertools.combinations(numeros_lotofacil, 15))

# Criar um DataFrame para armazenar as combinações
df_combinacoes = pd.DataFrame(todas_combinacoes, columns=[f"D.{i}" for i in range(1, 16)])

# Exibir as primeiras linhas do DataFrame
print(df_combinacoes.head())

# Defina a função para formatar os números
def formatar_numero(numero):
    return f'{numero:.2f}'

# Configure a opção de formatação de ponto flutuante do pandas
pd.options.display.float_format = formatar_numero

# Exiba as estatísticas descritivas do DataFrame
print(df_combinacoes.describe())
```

```python
df = df.drop(columns=['Data','Ganhadores','Concurso'])
df.info()
```

```python
# Criar características dos números sorteados
df['Soma_Numeros'] = df.loc[:, 'D. 1':'D. 15'].sum(axis=1)  # Soma dos números sorteados
df['Media_Numeros'] = df.loc[:, 'D. 1':'D. 15'].mean(axis=1)  # Média dos números sorteados
df['Pares'] = df.loc[:, 'D. 1':'D. 15'].apply(lambda row: sum(1 for num in row if num % 2 == 0), axis=1)  # Quantidade de números pares
df['Impares'] = df.loc[:, 'D. 1':'D. 15'].apply(lambda row: sum(1 for num in row if num % 2 != 0), axis=1)  # Quantidade de números ímpares

# Exemplo de criação de uma nova característica com base na soma dos dois primeiros números sorteados
df['Soma_Primeiros_Numeros'] = df['D. 1'] + df['D. 2']

# Função para contar números primos
def contar_primos(numeros):
    primos = 0
    for num in numeros:
        if num > 1:
            for i in range(2, num):
                if (num % i) == 0:
                    break
            else:
                primos += 1
    return primos

# Contagem de números primos por jogo
df['Numeros_Primos'] = df.loc[:, 'D. 1':'D. 15'].apply(contar_primos, axis=1)

# Função para encontrar a maior sequência de números em cada jogo
def maior_sequencia(numeros):
    max_sequencia = 0
    sequencia_atual = 0
    for i in range(len(numeros) - 1):
        if numeros[i + 1] - numeros[i] == 1:
            sequencia_atual += 1
        else:
            if sequencia_atual > 0:
                max_sequencia = max(max_sequencia, sequencia_atual)
            sequencia_atual = 0
    return max_sequencia + 1  # Adicionando 1 porque estamos considerando a sequência atual como não sequencial no loop

# Maior sequência de números por jogo
df['Maior_Sequencia'] = df.loc[:, 'D. 1':'D. 15'].apply(lambda row: maior_sequencia(sorted(row)), axis=1)

# Adicionar características adicionais ao conjunto de dados
df['Desvio_Padrao'] = df.loc[:, 'D. 1':'D. 15'].std(axis=1)  # Desvio padrão dos números sorteados
df['Mediana'] = df.loc[:, 'D. 1':'D. 15'].median(axis=1)  # Mediana dos números sorteados
df['Moda'] = df.loc[:, 'D. 1':'D. 15'].mode(axis=1).iloc[:, 0]  # Moda dos números sorteados

# Adicionar a variância dos números sorteados
df['Variância'] = df.loc[:, 'D. 1':'D. 15'].var(axis=1)  # Variância dos números sorteados

# Adicionar um histograma como uma característica
df['Histograma'] = df.loc[:, 'D. 1':'D. 15'].apply(lambda row: [row.tolist().count(i) for i in range(1, 26)], axis=1)  # Histograma dos números sorteados

# Exibir as primeiras linhas do DataFrame com as novas características
print(df.head())

```
